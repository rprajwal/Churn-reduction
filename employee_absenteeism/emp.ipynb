# Outlier analysis

```python
#continuous variables
cont=['ID','Transportation expense','Distance from Residence to Work','Service time','Age','Work load Average/day ','Hit target',
      'Weight','Height','Body mass index']
#categorical variables
categ=['Reason for absence','Month of absence','Day of the week','Seasons','Disciplinary failure','Education','Son','Social drinker',
      'Social smoker','Pet']
```

```python
#Boxplots to find outliers in continuous variables
for i in cont:
    plt.boxplot(d[i])
    plt.title(i)
    plt.show()
```

```python
#Removing outliers from 'Work load Average/day' variable
q75=d['Work load Average/day '].describe()[6]
q25=d['Work load Average/day '].describe()[4]
iqr=q75-q25
maxi=q75+(iqr*1.5)
d['Work load Average/day '].loc[d['Work load Average/day ']>maxi]=maxi
```

# Model development

```python
#sampling the data
from sklearn.cross_validation import train_test_split
test,train=train_test_split(d,test_size=0.8)
test.shape
```

# Decision tree

```python
#Decision tree
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(max_depth=2).fit(train.iloc[:,:19],train.iloc[:,19])
```

```python
#predicting values
pv=dt.predict(test.iloc[:,:19]).round()
```

```python
#calculating mape
def mape(x,y):
    mape=np.mean(np.abs((x-y)/x))*100
    return mape
```

```python
mape(test.iloc[:,19],pv)
```

```python
#Since we have 0 values in our target variable, mape will not give the proper results
#Calculating mae
from sklearn.metrics import mean_absolute_error
print('MAE = ',mean_absolute_error(test.iloc[:,19],pv))
```

# Random forest

```python
#importing rf library
from sklearn.ensemble import RandomForestRegressor

#applying rf model on train data
rf=RandomForestRegressor(n_estimators=500).fit(train.iloc[:,:19],train.iloc[:,19])

#predicting values
p=rf.predict(test.iloc[:,:19]).round()

#Calculating mae
print('MAE = ',mean_absolute_error(test.iloc[:,19],p))
```

# Linear regression

```python
#importing library
from sklearn.linear_model import LinearRegression

#applying linear regression on train data
lr=LinearRegression()
lr.fit(train.iloc[:,:19],train.iloc[:,19])

#predicting values
p=lr.predict(test.iloc[:,:19]).round()

#Calculating mae
print('MAE = ',mean_absolute_error(test.iloc[:,19],p))
```

# KNN

```python
#importing library
from sklearn.neighbors import KNeighborsRegressor

#applying KNN on train data
km=KNeighborsRegressor(n_neighbors=3).fit(train.iloc[:,:19],train.iloc[:,19])

#predicting values
p=km.predict(test.iloc[:,:19]).round()

#Calculating mae 
print('MAE = ',mean_absolute_error(test.iloc[:,19],p))
```
